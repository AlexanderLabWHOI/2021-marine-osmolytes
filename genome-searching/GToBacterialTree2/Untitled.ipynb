```python
import pandas as pd
```

```python
order = []
with open('tree-order', 'r') as f:
    for line in f:
        order.append(line.strip())
data  = pd.read_csv('../predict_synthesis/predicted-bacterial-osmolyte-synthesis.tsv', sep = '\t', header=[0,1], index_col=0)
```

```python
SYNTH = data.xs('SYNTHESIS', level=1, drop_level=False, axis=1)
SYNTH.columns = SYNTH.columns.droplevel(1)
SYNTH = SYNTH.loc[order]
SYNTH.to_csv('TREE_SYNTH.tsv', sep = '\t')
```

```python
BD = data.xs('BREAKDOWN', level=1, drop_level=False, axis=1)
BD.columns = BD.columns.droplevel(1)
BD = BD.loc[order]
BD.to_csv('TREE_BD.tsv', sep = '\t')
```

```python
TRAN = data.xs('TRANSPORT', level=1, drop_level=False, axis=1)
TRAN.columns = TRAN.columns.droplevel(1)
TRAN = TRAN.loc[order]
TRAN.to_csv('TREE_TRAN.tsv', sep = '\t')
```
